```python
import import_ipynb
import numpy as np, pandas as pd
import cv2, time, os
```

```python
face_detect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

# Loading or Generating Data to Train the Model:

```python
path = 'data_for_training'
dir = os.listdir(path)

if len(dir) != 0:
    input_images = []
    targets = []
    for images in dir:
        filename = path + '/' + images
        face = cv2.imread(filename)
        gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        input_images.append(gray)
        targets.append(int(images.split('-')[1].split('.')[0]))
    print("Data loaded Successfully!")
    
else:
    print("Data unavailable for training. Scan of your face needed.")
    time.sleep(0.5)
    members = int(input("Mention the no. of members: "))
    source = cv2.VideoCapture(0)
    input_images = []
    targets = []
    counter = 0
    for i in range(members):
        no_of_images = 0
        input('Press enter to start capturing')
        print("Get ready, scanning your face in 5 seconds.")
        time.sleep(5)
        while True:
            _, image = source.read(0)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            # detecting faces
            detected_faces = face_detect.detectMultiScale(gray, 1.2, 9)
            
            if len(detected_faces) > 1:
                continue
            else:
                pass
            
            for bbox in detected_faces:
                x,y,w,h = bbox

                # croping the faces for training the model
                cropped_image = cv2.resize(gray[y:y+h,x:x+w],(100,100))
                input_images.append(cropped_image)
                targets.append(counter)
                
                # Saving input_images for future use
                no_of_images += 1
                path = 'data_for_training'
                cv2.imwrite(os.path.join(path , str(no_of_images) + '-' + str(counter)+'.jpg'), cropped_image)

                # Formatting image for display
                cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 3)
                cv2.putText(image, "Image No : " + str(no_of_images), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 100, 100), 2)
                cv2.imshow("Scanning your face", image)
                cv2.waitKey(1)
            if no_of_images == 500:
                break
        
        counter += 1
        
        cv2.destroyAllWindows()
    source.release()
```

# Preparing the Data for Training the Neural Network:

```python
input_images_normalized = np.array(input_images)/255.0
input_images_normalized
```

```python
input_images_normalized_reshaped = input_images_normalized.reshape(len(input_images_normalized),100,100,1)
input_images_normalized_reshaped
```

```python
# For One-Hot Encoding:

from keras.utils import np_utils
targets = np_utils.to_categorical(targets)
targets
```

```python
from sklearn.model_selection import train_test_split

X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(input_images_normalized_reshaped, targets)
```

# Building and Training the Neural Network:

```python
from keras.models import Sequential
from keras.layers import Conv2D, Dense, Activation, Flatten
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()

model.add(Conv2D(300, (3, 3), input_shape = (100, 100, 1)))
model.add(Activation('relu'))

model.add(Conv2D(100, (3,3)))
model.add(Activation('relu'))

model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(len(targets[0]), activation='softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['categorical_accuracy'])
```

### Defining Checkpoint and Training the Model:

```python
cp = ModelCheckpoint('best_one', verbose=0, save_best_only=True)
model.fit(X_TRAIN, Y_TRAIN, epochs = 4, callbacks=[cp], validation_split=0.2)
```

```python
from keras.models import load_model
model = load_model('best_one')
```

# Defining function for Sending SMS:

```python
import conf
from twilio.rest import Client
```

#### Note: conf is the file containing personal data required for sending SMS

```python
def send_twilio_alert(alert):
    account_sid = conf.SID
    auth_token = conf.AUTH_TOKEN
    TO_NUMBER = conf.TO_NUMBER
    FROM_NUMBER = conf.FROM_NUMBER
    client = Client(account_sid, auth_token) 
 
    message = client.messages.create(to = TO_NUMBER,from_ = FROM_NUMBER, body = alert)
    print(message.sid)
```

# Defining function for Alerting on Telegram:

```python
import requests, json
```

```python
def send_telegram_alert(alert):
    url = "https://api.telegram.org/bot" + conf.telegram_bot_id + "/sendMessage"
    data = {
        "chat_id": conf.telegram_chat_id,
        "text": alert
    }
    try:
        response = requests.request(
            "POST",
            url,
            params=data
        )
        telegram_data = json.loads(response.text)
        if telegram_data["ok"] == True:
            return "Alert Raised."
    except Exception as e:
        print("An error occurred in sending the alert message via Telegram")
        print(e)
        return False
```

# Initiating Surveillance:

```python
from IPython.display import clear_output
```

```python
source = cv2.VideoCapture(0)
while True:
    prob_is_me = []
    _, image = source.read()
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # detecting faces
    detected_faces = face_detect.detectMultiScale(gray, 1.2, 9)
    
    if len(detected_faces) == 0:
        continue
    else:
        pass
    
    for each_face in detected_faces:

        x, y, w, h = each_face
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 3)
        # Cropping the image
        crop = gray[y:y+h, x:x+w]
        cropped_image = cv2.resize(crop, (100, 100))
        
        # Normalising and reshaping image
        normalized_image = cropped_image/255
        reshaped_face = np.reshape(normalized_image, (1, 100, 100, 1))
        
        # Displaying captured image
        cv2.imshow("Scanning your face", image)
        
        #Predicting and collecting results in a list
        result = model.predict(reshaped_face)[0]

        prob_is_me.append(result[0])
    print(prob_is_me)
   

    if max(prob_is_me) < 0.90:
        print("Someone unrecognized detected! Sending alerts.")
        #send_twilio_alert("Caution! Someone is trying to trespass the security. \n"
        #                 +"Number of intruders : " + str(len(prob_is_me)))
        telegram_status = send_telegram_alert(alert = "Caution! Someone is trying to trespass the security. \n"
                                             +"Number of intruders : " + str(len(prob_is_me)))
        print("Telegram Status: ", telegram_status)
        time.sleep(5)
    else:
        #time.sleep(7)
        pass
    
    time.sleep(0.5)
    clear_output(wait=True)   
    
    key = cv2.waitKey(1)
    if key == 27:
        break
    else:
        pass
cv2.destroyAllWindows()
source.release()
```
